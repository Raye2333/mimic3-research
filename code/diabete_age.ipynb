```python
age_mat['AGE: 0-10'] = [0]*pat_num
age_mat['AGE: 11-20'] = [0]*pat_num
age_mat['AGE: 21-30'] = [0]*pat_num
age_mat['AGE: 31-40'] = [0]*pat_num
age_mat['AGE: 41-50'] = [0]*pat_num
age_mat['AGE: 51-60'] = [0]*pat_num
age_mat['AGE: 61-70'] = [0]*pat_num
age_mat['AGE: 71-80'] = [0]*pat_num
age_mat['AGE: 81-89'] = [0]*pat_num
age_mat['AGE: 90+'] = [0]*pat_num
```

```python
age_mat['AGE: 90+'] = list(df_age['OVER_80'])
```

```python
age_mat.at[0, 'AGE: 0-10']
```

```python
for i in range(pat_num):
    this_age = age_mat.iloc[i, 1]
    if this_age <= 0:
        age_mat.at[i, 'AGE: 0-10'] = 1
    elif this_age > 0 and this_age <= 10:
        age_mat.at[i, 'AGE: 0-10'] = 1
    elif this_age > 10 and this_age <= 20:
        age_mat.at[i, 'AGE: 11-20'] = 1
    elif this_age > 20 and this_age <= 30:
        age_mat.at[i, 'AGE: 21-30'] = 1
    elif this_age > 30 and this_age <= 40:
        age_mat.at[i, 'AGE: 31-40'] = 1
    elif this_age > 40 and this_age <= 50:
        age_mat.at[i, 'AGE: 41-50'] = 1
    elif this_age > 50 and this_age <= 60:
        age_mat.at[i, 'AGE: 51-60'] = 1
    elif this_age > 60 and this_age <= 70:
        age_mat.at[i, 'AGE: 61-70'] = 1
    elif this_age > 70 and this_age <= 80:
        age_mat.at[i, 'AGE: 71-80'] = 1
    elif this_age > 80 and this_age <= 89:
        age_mat.at[i, 'AGE: 81-89'] = 1
    print(str(age_mat.iloc[i, 0]) + ': ' + 'Age ' + str(this_age))
```
