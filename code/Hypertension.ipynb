### 3.14 Get Ages

```python
hyper3 = hyper2.copy()
```

```python
# Get the LAST time the patient comes to hosptalization
admin_age = admin.drop_duplicates(subset = 'SUBJECT_ID', keep = 'last')
ADMITTIME = list(admin_age[admin_age['SUBJECT_ID'].isin(pat_name)].sort_values(by = 'SUBJECT_ID')['ADMITTIME'])
DOB = list(patient[patient['SUBJECT_ID'].isin(pat_name)].sort_values(by = 'SUBJECT_ID')['DOB'])
```

```python
def AgeDelta(dob, doa):
    dob_date = dt.datetime(int(dob[0:4]), int(dob[5:7]), int(dob[8:10]))
    doa_date = dt.datetime(int(doa[0:4]), int(doa[5:7]), int(doa[8:10]))
    admit_age = int((doa_date - dob_date).days/365)
    if admit_age > 90:
        above_90 = 1
        admit_age = 90
    else:
        above_90 = 0
    ordinal_age = (admit_age//10) + 1
    return [admit_age, above_90, ordinal_age]
```

```python
ADMIT_AGE = []
ABOVE_90 = []
ORDINAL_AGE = []
for i in range(unique_hyper):
    dob = DOB[i]
    doa = ADMITTIME[i]
    tri = AgeDelta(dob, doa)
    ADMIT_AGE.append(tri[0])
    ABOVE_90.append(tri[1])
    ORDINAL_AGE.append(tri[2])
    print('Patient No.', str(i + 1), ', Aged', str(tri[0]))
```

```python
hyper3['ADMIT_AGE'] = ADMIT_AGE
hyper3['ABOVE_90'] = ABOVE_90
hyper3['ORDINAL_AGE'] =ORDINAL_AGE
```
