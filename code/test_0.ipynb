```python
# Diagnoses.csv
diag_df = pd.read_csv("DIAGNOSES_ICD.csv")
diag_df = (diag_df.sort_values("SUBJECT_ID")).drop(columns = "ROW_ID")
diag_df = diag_df.reset_index(drop = True)
diag_df.head(20)
```

```python
icd_code = pd.read_csv("D_ICD_DIAGNOSES.csv")
icd_code = (icd_code.drop(columns = "ROW_ID")).sort_values("ICD9_CODE")
icd_code = icd_code.reset_index(drop = True)
icd_code.head(20)
```

```python
# data prep and merging
pat_df = pd.read_csv("PATIENTS.csv")
pat_df = (pat_df.drop(columns = "ROW_ID")).sort_values("SUBJECT_ID")
pat_df = pat_df.reset_index(drop = True)
pat_df.head(20)
```

```python
# Merge dataframes
diag_icd9 = diag_df.copy()
diag_icd9.head()
```

```python
len(diag_df)
```

```python
diag_icd9['SHORT_TITLE'] = 0
diag_icd9['LONG_TITLE'] = 0
for i in range(len(diag_icd9)):
    this_code = diag_icd9['ICD9_CODE'][i]
    #print(this_code)
    icd_sub = icd_code[icd_code['ICD9_CODE'] == this_code]
    if len(icd_sub) > 0:
        diag_icd9['SHORT_TITLE'][i] = icd_sub.iloc[0]['SHORT_TITLE']
        diag_icd9['LONG_TITLE'][i] = icd_sub.iloc[0]['LONG_TITLE']   
    else:
        diag_icd9['SHORT_TITLE'][i] = 'N/A'
        diag_icd9['LONG_TITLE'][i] = 'N/A'
diag_icd9
```
